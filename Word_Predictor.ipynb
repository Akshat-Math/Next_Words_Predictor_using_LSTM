```python
import tensorflow as tf
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([dataset])
```

```python
# total number of words
num_words = len(tokenizer.word_index)
num_words
```

```python
# assigned an index to every word
tokenizer.word_index
```

```python
# sentence by sentence dataset ko breakdown kar diya
for sentence in dataset.split("\n"):
  print(sentence)
```

```python
#now convert words to numbers
for sentence in dataset.split("\n"):
  print(tokenizer.texts_to_sequences([sentence])[0])

# [0] isliye kyuki iski binda [[]] double list create ho rahi thi
# see jaise first word in dataset is 'Pride' jisko index 9 assign hua hai, and yaha wo first hai dekho
```

```python
input_sequences = []
for sentence in dataset.split("\n"):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
# sab rows ka size barabar karna hai kyuki supervised mai sabka shape barabar hota hai
# so apply zero padding
# so pehle max length waala sentence pata karlo aur uske hisaab se padding kar do
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding = 'pre')
```

```python
padded_input_sequences
```

```python
# so rows(sentences) sab li hai i.e [:]
# but columns mai se last waala hata diya hai i.e [:-1]
x = padded_input_sequences[:,:-1]
x
```

```python
# jo last waala tha wo y mai add ho gya
y = padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y,num_classes = num_words+1)

# +1 isliye kyuki one hot encoding humesha 0 se start hota hai
# and apne number of words 1 se start hai
# so we miss last waala word if num_words daal dete hai to

print(y.shape)
print(y[0])
```

```python
# Architecture
```

```python
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
```

```python
# so har row mai we have 101 elements as apan ne 1 remove kar diay hai
# so we get 100 as input
max_len
```

```python
model.add(Embedding(260,100,input_length = 100))
model.add(LSTM(150)) # 150 is just number of nodes, apne hisaab se
model.add(Dense(260,activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs = 100)
```

```python
# ye training accuracy hai and i think ki validation itti acchi nahi hogi
```

```python
text = 'Pride'

#tokenize
token_text = tokenizer.texts_to_sequences([text])[0]
#padding
padded_token_text = pad_sequences([token_text], maxlen=100, padding = 'pre')
print(padded_token_text)

# predict
model.predict(padded_token_text)
model.predict(padded_token_text).shape
```

```python
import numpy as np
```

```python
position = np.argmax(model.predict(padded_token_text))
#means 155th index pe jo word hai wo predicted word hai
```

```python
for word,index in tokenizer.word_index.items():
  if index == position:
    print(word)
```

```python
text = "Pride and"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(1)
```
